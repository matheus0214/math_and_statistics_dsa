**Exercício 1: Soma de Vetores**

Descrição:  Escreva  uma  função  em  Python  que  some  dois  vetores.  Suponha  que  os vetores  têm  o  mesmo  tamanho.  Resolva  o  exercício  de  forma  diferente  do  que  foi mostrado nas aulas.

```python
def sum_vectors(vec1: list[float], vec2: list[float]) -> list[float]:
    """
    Sum two vectors with the same length and return the result vector.

    Args:
        vec1 (list[float]): first vector
        y (list[float]): second vectors

    Result:
        list[float]: result vector
    """
    new_vec: list[float] = []

    for i in range(len(vec1)):
        new_vec.append(vec1[i] + vec2[i])

    return new_vec

res: list[float] = sum_vectors([1, 2, 3], [4, 5, 6])
print(res)
```

**Exercício 2: Produto Escalar**

Descrição:  Implemente  uma  função  que  calcule  o  produto  escalar  entre  dois  vetores. Resolva o exercício de forma diferente do que foi mostrado nas aulas.

```python
def calc_dot_product(v1, v2) -> float:
    return sum([x * y for x, y in zip(v1, v2)])
    
res = calc_dot_product([1, 3, -5], [4, -2, -1])
print(res)
```

**Exercício 3: Norma de um Vetor**

Descrição: Escreva uma função que calcule a norma (ou comprimento) de um vetor.

```python
import math

def calc_norm(x):
    return math.sqrt(sum(v**2 for v in x))

calc_norm([3, -4, 2]), calc_norm([2, -3, 1])
```

**Exercício 4: Ângulo entre Dois Vetores**

Descrição: Implemente uma função que calcule o ângulo (em graus) entre dois vetores.

```python
text1 = "abc"
text2 = "abd"

t1_ord = [ord(x) for x in text1]
t2_ord = [ord(x) for x in text2]

print(t1_ord, t2_ord)
```

```python
import math

def get_angle(v1, v2):
    dividend = 0
    
    a_sum = 0
    b_sum = 0

    for i in range(len(v1)):
        # scalar product
        dividend += (v1[i] * v2[i])

        # norm of vectors
        a_sum += v1[i]**2
        b_sum += v2[i]**2
    
    res: float = math.acos(dividend / round(math.sqrt(a_sum) * math.sqrt(b_sum)))

    return 180 * res / math.pi

result = get_angle([3, 4], [1, 0])

print(f"{result:.2f}°")
```

**Exercício 5: Projeção de um Vetor sobre Outro**

Descrição: Escreva uma função que encontre a projeção de um vetor sobre outro vetor.

```python
import math

def find_projection(a: list[float], b: list[float]) -> float:
    dividend = 0
    divisor = 0

    for i in range(len(a)):
        dividend += a[i] * b[i]
        divisor += b[i]**2

    return dividend / abs(math.sqrt(divisor))


result: float = find_projection([3, 4], [1, 0])

print(f"{result:.2f}")
```
