## Representação Gráfica de Funções

Vamos criar um exemplo com a função quadrática f(x) = x ^ 2

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def f(x):
    return x ** 2
```

```python
# Cria um conjunto de valores aleatórioas de x (por exemplo, de -10 a 10)
x = np.linspace(-10, 10, 40)
```

```python
y = f(x)
```

```python
print(x)
```

```python
plt.figure(figsize=(6, 4))
plt.plot(x, y, label = 'f(x) = x²')

plt.title('Gráfico da Função Quadrática')
plt.xlabel('x')
plt.ylabel('f(x)')

plt.legend()

plt.grid(True)
plt.show()
```

### Exemplo de uso da transformação de Log em Data Science

```python
# Gerando dados de exemplo: preços de casas
np.random.seed(0)                  # média   desvio padrão  qtd.
house_prices = np.random.lognormal(mean = 13, sigma = 0.6, size = 1000)

# Aplicando a transformação logarítimica
house_prices_log = np.log(house_prices)

# Criando os gráficos para comparação
plt.figure(figsize = (8, 4))

# Gráfico dos dados originais
plt.subplot(1, 2, 1)
plt.hist(house_prices, bins = 30, color = 'blue', alpha = 0.7)
plt.title('Distribuição dos Preços das Casas')
plt.xlabel('Preço')
plt.ylabel('Frequência')

# Gráfico dos dados transformados
plt.subplot(1, 2, 2)
plt.hist(house_prices_log, bins = 30, color = 'red', alpha = 0.7)
plt.title('Distribuição Logarítmica dos Preços das Casas')
plt.xlabel('Log do Preço')
plt.ylabel('Frequência')

plt.tight_layout()
plt.show()
```
