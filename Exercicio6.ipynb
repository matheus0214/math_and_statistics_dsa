**Exercício 2: Multiplicação de Matrizes**

Implemente  uma  função  em  Python  que  multiplica  duas  matrizes.  Verifique  se  as dimensões  das  matrizes  são  compatíveis  para  a  multiplicação  e  retorne  a  matriz  resultante. Apresente a solução com e sem o uso do NumPy. Pratique suas habilidades de programação ao mesmo tempo que estuda Matemática

```python
import numpy as np

def multi_matrix(a, b):
    # verify is tha matrix had the same size
    a_rows = len(a)
    b_rows = len(b)

    a_cols = set([len(a[i]) for i in range(a_rows)])
    b_cols = set([len(b[i]) for i in range(b_rows)])
    

    if len(a_cols) > 1 or len(b_cols) > 1:
        raise ValueError("cols should be the same length")

    a_cols = list(a_cols)[0]
    b_cols = list(b_cols)[0]

    # (ar,ac) x (br,bc) -> ac == br
    if a_cols != b_rows:
        raise ValueError("unable to calculate the matriz multi")

    # result
    # (an,am) x (bn,bm) -> an,bm
    total = []

    for i in range(a_rows):
        r = []
        k = 0
        while k < b_cols:
            s = 0
            for j in range(a_cols):
                s += (a[i][j] * b[j][k])
            
            k += 1
            r.append(s)

        total.append(r)

    return total


print(multi_matrix(m1, m2))
print(np.matmul(m1, m2))
```

**Exercício 3: Transposição de Matriz**

Crie um programa que calcula a transposta de uma matriz. A matriz deve ser inserida pelo usuário e o programa deve exibir a matriz transposta. Apresente a solução com e sem o uso do NumPy. Pratique suas habilidades de programação ao mesmo tempo que estuda Matemática.

```python
import numpy as np

def transpose(a):
    rows = len(a)
    cols = list(set([len(x) for x in a]))

    if len(cols) > 1:
        raise ValueError("invalid cols")

    cols = cols[0]
    shape = (rows, cols)
    
    T = []
    for i in range(cols): # new lines
        n_row = []
        for j in range(rows): # new columns
            n_row.append(a[j][i])

        T.append(n_row)

    return T


print(transpose(m1))
print(np.transpose(m1))
```
