### Exercício 1: Gráfico de Uma Função Linear

Crie um gráfico para a função y = 2x + 3 para valores de x no intervalo de 5 a 10.

```python
def linear_function(x):
    return 2 * x + 3

x = [x for x in range(5, 11)] # np.linspace(5, 10, 100)
y = [linear_function(value) for value in x]

plt.figure(figsize = (6, 4))

plt.plot(x, y, label = 'f(x) = 2x + 3')

plt.title('Gráfico da Função Linear')
plt.xlabel('X')
plt.ylabel('y = 2x + 3')
plt.grid(True)

plt.legend()
plt.show()
```

### Exercício 6: Resolver uma Equação Quadrática

Escreva um script em Python que resolve a equação x^2 − 5x + 6 = 0 e imprime as raízes.

```python
import cmath

def quadratic_function():
    a = 1
    b = -5
    c = 6
    
    delta = cmath.sqrt((b)**2 - (4 * a * c))
    x1 = (-b + delta) / (2*a)
    x2 = (-b - delta) / (2*a)
    
    return (x1, x2)


quadratic_function()
```

### Exercício 7: Equação Quadrática com Entrada do Usuário

Modifique o programa anterior para aceitar coeficientes a, b, e c de um usuário e resolver a equação quadrática correspondente.

```python
import cmath

def quadratic_function(a, b, c):
    delta = cmath.sqrt((b)**2 - (4 * a * c))
    x1 = (-b + delta) / (2*a)
    x2 = (-b - delta) / (2*a)
    
    return (x1, x2)

a = float(input('Informe o valor de a:'))
b = float(input('Informe o valor de b:'))
c = float(input('Informe o valor de c:'))

result = quadratic_function(a, b, c)

print(f'Resultado {result}')
```

### Exercício 8: Equação Quadrática com Discriminante

Para a equação quadrática ax^2 + bx + c = 0, calcule o discriminante e determine o número de raízes reais.

```python
def discriminant(a, b, c):
    return b**2 - 4*a*c

a = float(input('Informe o valor de a:'))
b = float(input('Informe o valor de b:'))
c = float(input('Informe o valor de c:'))

result = discriminant(a, b, c)

if result > 0:
    print('2 raizes possiveis')
elif result == 0:
    print('apenas uma raiz possivel')
else: # result < 0
    print('raizes imaginárias')
```

### Exercício 9: Cálculo de Logaritmos

Escreva um script que calcula o logaritmo de base 10 de um número fornecido pelo usuário. Certifique-se de tratar o caso em que o logaritmo não pode ser definido (números negativos e zero).

```python
# math.log10(50) 1.6989700043360187
# https://www.youtube.com/watch?v=bIDOTWy7gZk

from decimal import Decimal

def bounds(x, base = 10):
    result = int(x // base)
    boundary = 0
    while result > 0:
        result = int(result // base)
        boundary += 1

    return boundary, boundary + 1


def search(x, base, min_bound, max_bound): # binary search
    # min < exp < max
    max_bound = max(min_bound, max_bound)
    min_bound = min(min_bound, max_bound)
    
    exp = (max_bound + min_bound) / 2
    guess = base**exp
    precision = 0.0000001
    
    while abs(guess - x) > precision:
        exp = (max_bound + min_bound) / 2
        guess = base**exp
        
        if guess < x:
            min_bound = exp
        elif guess > x:
            max_bound = exp
    
    return exp



x = 5
base = 10

limits = bounds(x, base)

search(x, base, *limits)
```

### Exercício 10: Aplicação de Logaritmos

Utilize logaritmo para calcular o tempo necessário para que um investimento atinja um determinado valor, considerando uma taxa de crescimento contínua.

```python
import timeit
import numpy as np

tax = 0.01 # per month
amount = 1000
target_amount = 1500

def calc(amount, target, tax):
    # is considering each year
    one_year = 1
    return np.log10(target / amount) / np.log10(1 + (tax / one_year))


time_taken = timeit.timeit(lambda: calc(amount, target_amount, tax), number=10000)
print(f"Time taken: {time_taken} seconds")

# print(math.log(2) / math.log(1 + tax))
```

```python
# Implement calculation with while loop 
# just to verify the code performace
# IS NOT THE BEST APPROACH TO USE (PERFORMACE IS WORST AND IS NOT SO PRECISE AS ABOVE IMPLEMENTATION)

import timeit

tax = 0.01 # per month
amount = 1000
target_amount = 1500

def calc_loop(amount, target, tax):
    acc = amount
    month = 0

    while acc < target:
        acc += acc * tax
        month += 1

    return month

time_taken = timeit.timeit(lambda: calc_loop(amount, target_amount, tax), number=10000)
print(f"Time taken: {time_taken} seconds")
```
