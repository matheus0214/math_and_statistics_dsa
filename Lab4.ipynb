### Geometria de vetores e representação em Python

```python
u = np.array([-2, 3])
```

```python
print(u)
```

```python
# Exemplo 1

# Criar a figura
plt.figure(figsize=(5,5))

# Vetor u
vector1 = plt.arrow(0, 0, u[0], u[1], head_width=.3, width=.1, color='blue', length_includes_head=True)

# Ponto de início do vetor
plt.plot(0, 0, 'ko', markerfacecolor='yellow', markersize=8)

# Formação do plot
plt.grid(linestyle='--', linewidth=.6)
plt.axis('square')
plt.axis([-4, 4, -4, 4])
plt.legend([vector1], ['u'])
plt.title(r'Vetor $\mathbf{u}$')
plt.savefig('images/graph_01.png', dpi=300)

plt.show()
```

```python
# Exemplo 2

# Cria a figura
plt.figure(figsize=(5, 5))

# Criar um array de vetores
vectors = np.array([ [3, -2], [-1, 4], [2, -3] ])
print('Vetores')
print(vectors)

# Cria um array de vetores para as origens
origins = np.array([ [0, -1, -2], [0, -2, -1] ])
print('\nOrigens')
print(origins)

# Plot
plt.quiver(*origins, vectors[:, 0], vectors[:, 1], color=['r', 'g', 'b'], scale=15)

# Escala
plt.axis([-5, 5, -5, 5])

plt.savefig('images/graph_02.png', dpi=300)

plt.show()
```
